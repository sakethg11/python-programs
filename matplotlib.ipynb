```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([98,87,88,75,83,50])
mylabels=['chemistry','maths','c language','physics','evpr','beee']
myexplode=[0.2,0,0,0,0,0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
